```python
import pandas as pd
import os

def process_folder_with_target_no_header(folder_name, base_path):
    X_list = []
    y_list = []

    folder_path = os.path.join(base_path, folder_name)
    
    files = os.listdir(folder_path)
    
    print(f"Files in {folder_name} folder: {files}")

    for file in files:
        file_path = os.path.join(folder_path, file)
        
        if folder_name == 'Gain' and file.startswith('gain'):
            try:
                X_data = pd.read_excel(file_path, header=[0, 1])  
                X_list.append(X_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
        
        elif folder_name == 'Gain' and file.startswith('ygain'):
            try:
                y_data = pd.read_excel(file_path, header=None)  
                y_list.append(y_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

        elif folder_name == 'Offset' and file.startswith('offset'):
            try:
                X_data = pd.read_excel(file_path, header=[0, 1])  
                X_list.append(X_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

        elif folder_name == 'Offset' and file.startswith('y'):
            try:
                y_data = pd.read_excel(file_path, header=None) 
                y_list.append(y_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

        elif folder_name == 'Stuck-at' and file.startswith('stuck'):
            try:
                X_data = pd.read_excel(file_path, header=[0, 1])  
                X_list.append(X_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

        elif folder_name == 'Stuck-at' and file.startswith('y'):
            try:
                y_data = pd.read_excel(file_path, header=None)  
                y_list.append(y_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

      
        elif folder_name == 'Out of bound' and file.startswith('out'):
            try:
                X_data = pd.read_excel(file_path, header=[0, 1])  
                X_list.append(X_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

       
        elif folder_name == 'Out of bound' and file.startswith('y'):
            try:
                y_data = pd.read_excel(file_path, header=None)  
                y_list.append(y_data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

  
    if not X_list:
        print(f"No feature files found in {folder_name}")
    if not y_list:
        print(f"No target files found in {folder_name}")

   
    if X_list:
        X_combined = pd.concat(X_list, axis=0, ignore_index=True)
    else:
        X_combined = pd.DataFrame()  
    if y_list:
        y_combined = pd.concat(y_list, axis=0, ignore_index=True)
    else:
        y_combined = pd.DataFrame()  
    if not X_combined.empty:
        X_combined.replace(1, 'normal', inplace=True)
        X_combined.replace(-1, folder_name, inplace=True)
    
    if not y_combined.empty:
        y_combined.replace(1, 'normal', inplace=True)
        y_combined.replace(-1, folder_name, inplace=True)

    if not X_combined.empty and not y_combined.empty:
        combined_data = pd.concat([X_combined, y_combined], axis=1)
    else:
        combined_data = pd.DataFrame()
    return combined_data

def combine_all_folders_with_targets(base_path):
   
    folder_names = ['Gain', 'Out of bound', 'Offset', 'Stuck-at']
    
    all_combined_data = []

    for folder in folder_names:
        combined_data = process_folder_with_target_no_header(folder, base_path)
        if not combined_data.empty:
            all_combined_data.append(combined_data)

    if all_combined_data:
        final_combined_data = pd.concat(all_combined_data, axis=0, ignore_index=True)
    else:
        final_combined_data = pd.DataFrame()  
    

    if not final_combined_data.empty:
        final_combined_data.to_csv(os.path.join(base_path, 'combined_data.csv'), index=False)
    
base_path = r"D:\Downloads\Network Fault Detection"

# Run the process
combine_all_folders_with_targets(base_path)
```
